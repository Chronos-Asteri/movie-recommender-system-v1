```python
from googletrans import Translator
translator = Translator()
detected_language = translator.detect('El Mariachi').lang
detected_language
```

```python
def detect_language(text):
    detected_language = translator.detect(text).lang
    return detected_language
```

```python
movies['title_language'] = movies['original_title'].apply(detect_language)
```
